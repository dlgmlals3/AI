```python
height = [155,160,165,172,177,180,185]
weight = [43,55,62,64, 67,74,90]
# 원의크기 크게
plt.scatter(height, weight, s=[100, 150, 200, 250, 300, 350, 400], c=['r']) 
plt.scatter(height, weight, s=[100, 150, 200, 250, 300, 350, 400], c=['r','g','b','k','m','y','c']) 

plt.show()
```

```python
# 표준체중(kg)=(신장(cm)-100)×0.85
# 비만도(%)=현재체중/표준체중×100
# 비만도가
# 90이하:저체중, (y,100) --> 노란색 , 크기
# 90∼110:정상,  (b ,150)
# 110∼120:과체중, (c, 200)
# 120이상:비만    (r ,300)
    
# 그래프각점에 비만도결과값(저체중, 정상,과체중, 비만)글자를
# 표시하고 원의 크기는  각 결과 값에 다른 크기로
# 그리시요.

result = []
color = []
size = []

def obesityFn(w, h):
    stdWeight = ( round((h - 100) * 0.85, 2) )
    heavy = round(w / stdWeight * 100, 2)
    if heavy >= 120:
        result.append('비만')
        color.append('r')
        size.append(300)
    elif 110 <= heavy < 120:
        result.append('과체중')
        color.append('c')
        size.append(200)
    elif 90 <= heavy < 110:
        result.append('정상')
        color.append('b')
        size.append(150)
    else:
        result.append('저체중')
        color.append('y')
        size.append(100)

height = [155,160,165,172,177,180,185]
weight = [43,55,62,64, 67,74,90]
for w, h in zip(weight, height):
    obesityFn(w, h)
    
plt.scatter(height, weight, size, color) 

for x,y,r in zip( height, weight, result):
    plt.text( x,y,r)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import folium

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x=['홍길동','이순신','임꺽정','이이']
y=[90,77,50,80]
plt.barh( x, y, color=['r','g','b','y'])
plt.show()
```

```python
xname =['홍길동','이순신','임꺽정','이이']
x = pd.Series([1,2,3,4]) 
kor=[90,77,50,80]
eng=[60,20,50,20]
plt.xticks(x+0.2, xname)
plt.bar( x, kor, width=0.4)
# x + 0.4 --> eliment 연산을 위해 x를 리스트가 아닌 Series로 변경
plt.bar( x+0.4, eng, width=0.4) # 임꺽정 kor 점수를 가림..
plt.show()
```

```python
xname =['홍길동','이순신','임꺽정','이이']
x = pd.Series([1,2,3,4]) 
kor=[90,77,50,80]
eng=[60,20,50,20]
plt.xticks(x, xname)
plt.bar( x, kor)
# bottom을 쓰면 국어점수를 깔고 위로 올려라
# stacked bar : dataframe plot ( stacked=True )
plt.bar( x, eng, bottom=kor) 
plt.legend(['국어', '영어'])

for a, k,e in zip(x, kor, eng):
    plt.text(a, k/2, f'{k}점', va='center', ha='center')
    plt.text(a, k+e/2, f'{e}점', va='center', ha='center')
    
plt.show()
```

```python
# 참고

### stacked bar : daframeplot( stacked=True)
data =  {'eng':[10,30,50,70],
         'kor':[20,40,60,80],
         'math':[90,50,20,70]}
df = pd.DataFrame(data, index=['a','b','c','d'] )
df.plot( kind='bar',stacked=True)
```

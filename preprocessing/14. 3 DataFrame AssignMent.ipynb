```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

```

```python
df = pd.read_csv('data/crime_in_Seoul.csv', index_col='관서명', encoding='euc-kr', thousands=',')
df
```

```python
# 1. 살인발생이 가장높은 관서명, 살인발생,살인검거를  출력하시요
df.nlargest(1, '살인 발생', keep='all')[['살인 발생', '살인 검거']]
```

```python
# 2. 관서별 절도발생,절도검거에 대한 바차트를 그리시요
df.plot(kind = 'bar', y=['절도 발생', '절도 검거'])
plt.show()
```

```python
df.insert(8, '절도 검거율', ((df['절도 검거'] / df['절도 발생'])*100).round(2))
df
# 3. 절도검거율 컬럼을 추가 하시요.
# df['절도 검거율'] = ((df['절도 검거'] / df['절도 발생'])*100).round(2)
# df
```

```python
# 4 가장 검거율이 좋은 Top 3 경찰서는 어디인가 
# (살인/강도/강간/절도/폭력)의 비율의 총합 / 5 기준치로 정한다. 
# 관서명 검거율 형태로 출력한다.
# > 발생보다 검거가 더큰 경우는 뭐지....

df['살인 검거율'] = df['살인 검거'] / df['살인 발생']
df['강도 검거율'] = df['강도 검거'] / df['강도 발생']
df['강간 검거율'] = df['강간 검거'] / df['강간 발생']
df['절도 검거율'] = df['절도 검거'] / df['절도 발생']
df['폭력 검거율'] = df['폭력 검거'] / df['폭력 발생']
df['검거율'] = (df['살인 검거율'] + df['강도 검거율'] + df['강간 검거율'] + df['절도 검거율'] + df['폭력 검거율']) / 5;

result = df.nlargest(3, '검거율', keep='all')[['검거율']].round(2)
result
```

```python
# 5 가장 안전한 Top 3 도시의 경찰서 출력 하라
# 불안전지수는 모든 범죄의 발생 빈도수 * 가중치의 총합
# 가중치 ( 살인 * 10, 강도 * 5, 그외 1)
# 안전한 경찰서는 불안전지수가 가장 낮은 경찰서

df['불안전 지수'] = (df['살인 발생'] * 10) + (df['강도 발생'] * 5) + df['강간 발생'] + df['절도 발생'] + df['폭력 발생']
print("가장 안전한 도시 Top 3")
df.nsmallest(3, '불안전 지수', keep='all')[['불안전 지수']].round(2)
```

```python
# 6  강간 발생과 폭력 발생의 상관관계를 구하라
df.plot(kind='scatter', x='강간 발생', y='폭력 발생')
print('상관 지수 : ', df.corr()['강간 발생']['폭력 발생'].round(2))
plt.show()
```

```python
# 7  강간 발생과 살인 발생의 상관관계를 구하라
df.plot(kind='scatter', x='강간 발생', y='살인 발생')
print('상관 지수 : ', df.corr()['강간 발생']['살인 발생'].round(2))
plt.show()
df[ ['강간 발생', '살인 발생']]
```

# Merge

```python
employee = [{'empno':1, 'ename':'kim', 'dept':1},
            {'empno':2, 'ename':'lee', 'dept':2},
            {'empno':3, 'ename':'park', 'dept':1},
            {'empno':4, 'ename':'song', 'dept':3},
            {'empno':5, 'ename':'min', 'dept':2}
           ]
dept=[{'dept':1, 'deptname':'관리직'},
      {'dept':2, 'deptname':'영업직'},
      {'dept':3, 'deptname':'개발직'}
     ]
info =[{'empno':1, 'addr':'서울시','phone':'010-1111-1111'},
       {'empno':3, 'addr':'부산시','phone':'010-2222-2222'},
       {'empno':2, 'addr':'광주시','phone':'010-3333-3333'},
       {'empno':5, 'addr':'광주시','phone':'010-4444-4444'},
       {'empno':4, 'addr':'광주시','phone':'010-5555-5555'}
      ]
emp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

### Merge(병합) : 반드시 공통 컬럼이 있어야함


```python
mdf = pd.merge(emp, dept)
mdf.drop(columns=['dept'], inplace=True)
mdf
```

```python
mdf = pd.merge(emp, dept, on='dept')
mdf.drop(columns=['dept'], inplace=True)
mdf
```

```python
mdf1 = pd.merge( mdf,info, on='empno')
mdf1.drop( columns=['empno'], inplace=True)
mdf1
```
